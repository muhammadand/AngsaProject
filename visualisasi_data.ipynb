```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Memuat data dari file CSV
file_path = 'Data_finalproject (1).csv'  # Ganti dengan path ke file CSV Anda
df = pd.read_csv(file_path)
```

```python
# Bar Chart - Jumlah komentar per pengguna
user_comment_count = df['Name'].value_counts()
plt.figure(figsize=(10,6))
user_comment_count.plot(kind='bar')
plt.title('Jumlah Komentar per Pengguna')
plt.xlabel('Username')
plt.ylabel('Jumlah Komentar')
plt.show()
```

```python
# Word Cloud - Kata-kata yang paling sering muncul dalam komentar
all_comments = ' '.join(df['Coment'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_comments)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud dari Komentar')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re
from wordcloud import STOPWORDS
```

```python
# Membersihkan dan Menggabungkan Semua Komentar
all_comments = ' '.join(df['Coment'])
```

```python
# Menghitung frekuensi kata
word_counts = Counter(all_comments.split())
```

```python
# Menghapus kata umum (STOPWORDS)
stopwords = set(STOPWORDS)
filtered_word_counts = {word: count for word, count in word_counts.items() if word not in stopwords}
```

```python
# Mendapatkan 10 kata yang paling sering muncul
most_common_words = dict(Counter(filtered_word_counts).most_common(10))

```

```python
# Mendapatkan 10 kata yang paling sering muncul
most_common_words = dict(Counter(filtered_word_counts).most_common(10))

```

```python
# Visualisasi dengan Bar Chart
plt.figure(figsize=(10,6))
plt.bar(most_common_words.keys(), most_common_words.values())
plt.title('10 Kata yang Paling Sering Muncul dalam Komentar')
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.show()
```
